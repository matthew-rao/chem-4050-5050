## Plot Overlap Integral Using Random Sampling

```python
#Import Necessary Packages
import monte_carlo
import matplotlib.pyplot as plt
```

```python
#Plot Random Sampling Overlap
plt.plot(monte_carlo.N_list, monte_carlo.randomsampling_sr)
#print(monte_carlo.randomsampling_sr)
plt.xscale("log")
plt.xlabel("Number of points sampled")
plt.ylabel("Integral Value")
plt.title("Overlap Integral of Two H 2pz orbitals (Random Sampling)")
plt.axhline(0.7366291166029653, color = "black", linestyle = "--")
plt.show()
```

### Discussion
Using our random sampling method, we see that the overlap integral of the 2pz orbitals converges to about 0.7366. We will subsequently compare this to our importance sampling method to determine the accuracy of our results.

# Overlap as Function of Separation Distance

```python
#Plot overlap as function of separation distance
plt.plot(monte_carlo.r_list, monte_carlo.separation_overlaps)
#print(monte_carlo.separation_overlaps)
plt.xlabel("Separation Distance")
plt.ylabel("Integral Value")
plt.title("Overlap Integral of Two H 2pz orbitals")
plt.axhline(0, color = "black", linestyle = "--")
plt.show()
```

### Change in Sign of S(R)
We observe that at a low separation distance, the S(R) has a positive value, but that this value reaches 0 at a separation of ~5.0. Additionally, after this, the S(R) value is negative and slowly converges to 0 as separation increases.
We can best explain this in terms of constructive and destructive interference. We know that the 2p orbital has 2 lobes, with one being positive and the other negative. At a small separation distance, the positive and negative ends of each 2pz orbital align, meaning that there is constructive interference. As separation increases, the positive end of one lobe begins to align with the negative lobe of the other orbital, creating destructive interference. At a distance at around 5, it means that there is just as much constructive interference as there is destructive interference. Further increases of the separation distance means that destructive interference dominates. As separation increases, the orbitals begin to lose their overlap altogether, meaning that while there is destructive interference present, its effects become smaller and smaller.

### Efficiency of Importance Sampling

We expect that importance sampling decreases the time that it takes to converge. This is because we are sampling from a probability distribution that resembles the distribution we are examining. In this case, we are utilizing the exponential continuous random variable. In this case, the samples are more heavily concentrated in regions of interest, allowing convergence to be reached faster. This is especially true in the example that we see in lecture 13. The simulation that we ran here, however, does not show the same. In fact, it converges at a factor that is 10 times slower than the random sampling. Possible explanations might include that the g(x,y,z) chosen wasn't a perfect match for our integrand. Another we might attribute to high variance in the system, which might explain the random fluctuations. However, on a more general basis, we typically expect the importance sampling to be more effective.

### Physical Significant of Overlap
The physical significance of our findings is understanding how covalent bonds form. We know that constructive interference leads to the formation of a covalent bond. In this sense, R directly affects the formation of covalent bonds. A small enough R ensures that constructive interference is present and allowing the bond to form. An increasing R means that destructive interference between orbital lobes is preventing this from occuring. And as R becomes sufficiently large, it indicates that the orbitals are too far away from each other for meaningful interactions to occur, thus preventing bonds from forming.
